```python
import pandas as pd
import numpy as np

# Load the CSV file
file_path = 'bok.csv'
df = pd.read_csv(file_path)

df.head()
```

```python
# Drop the columns that are not needed
cols = ['result', 'table', '_start', '_stop', '_measurement', 'pet_name']
df = df.drop(columns=cols)
df.head()
```

```python
# Change Column Names to be more simple
df.columns.values[0] = 'timestamp'
df.head(20)
```

```python
df.columns
df.columns = df.columns.str.strip()
```

```python
df['timestamp']
```

```python
# timestamp의 데이터를 띄어쓰기를 기준으로 나누어 date와 time 열을 만들기
df['date'] = df['timestamp'].str.split(' ').str.get(0)
df['times'] = df['timestamp'].str.split(' ').str.get(1)

# timestamp 열을 제거하기
df = df.drop(columns='timestamp')

df.head()
```

```python
#date을 datetime64 자료형으로 변환하기
df['date'] = pd.to_datetime(df['date'])
#date의 -를 .으로 바꾸기
df['date'] = df['date'].dt.strftime('%Y.%m.%d')


#time을 시:분 과 초로 나누어 time과 second 열을 만들기
df['time'] = df['times'].str.split(':').str.get(0) + ':' + df['times'].str.split(':').str.get(1)
df['second'] = df['times'].str.split(':').str.get(2)

#times 열을 제거하기
df = df.drop(columns='times')

#second에 +부터 끝까지의 문자열을 제거하기
df['second'] = df['second'].str.split('+').str.get(0)
df.head()
```

```python
print(df['date'].dtypes, df['time'].dtypes)
df.head()
```

```python
# 가속도 데이터를 g 단위로 변환 (기존 데이터가 이미 g 단위가 아닐 경우 필요)
g = 9.80665
df['ax'] = df['ax'].astype(float) * g
df['ay'] = df['ay'].astype(float) * g
df['az'] = df['az'].astype(float) * g

# Roll 계산
df['roll'] = np.arctan2(df['ay'], df['az'])

# Pitch 계산
df['pitch'] = np.arctan2(-df['ax'], np.sqrt(df['ay']**2 + df['az']**2))

# Yaw는 계산이 복잡하여 여기서는 생략합니다.

# Surge, Sway, Heave
df['surge'] = df['ax']
df['sway'] = df['ay']
df['heave'] = df['az']

df.head()
```

```python
print(df.dtypes)
```

```python
# date 칼럼을 string으로 변환
df['date'] = df['date'].astype(str)
# time 칼럼을 string으로 변환
df['time'] = df['time'].astype(str)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 시각화를 위한 데이터프레임 복사
df_plot = pd.DataFrame(df)

# date와 time을 합쳐서 datetime 형식으로 변환
df_plot['datetime'] = pd.to_datetime(df_plot['date'] + ' ' + df_plot['time'])

time_numeric = np.arange(len(df))

# 3D 플롯 생성
fig = plt.figure(figsize=(14, 10))
ax = fig.add_subplot(111, projection='3d')

# 3D 플롯에 데이터 추가
ax.plot(time_numeric, df['surge'], df['heave'], label='Surge vs Heave', color='blue')
ax.plot(time_numeric, df['sway'], df['heave'], label='Sway vs Heave', color='orange')

# 축 레이블 설정
ax.set_xlabel('Time (numeric)')
ax.set_ylabel('Surge/Sway')
ax.set_zlabel('Heave')

# 플롯 타이틀과 레전드 설정
ax.set_title('3D Visualization of Surge, Sway, and Heave over Time')
ax.legend()

# 플롯 표시
plt.show()
```
